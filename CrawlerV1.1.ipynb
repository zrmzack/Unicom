```python
def get_track(distance):

    track = []
    # 当前位移
    current = 0
    # 减速阈值
    mid = distance * 4 / 5
    # 计算间隔
    t = 0.1
    # 初速度
    v = 4
 
    while current < distance:
        if current < mid:
            # 加速度为正50
            a = 50
        else:
            # 加速度为负10
            a = -10
        # 初速度v0
        v0 = v
        # 当前速度v = v0 + at
        v = v0 + a * t
        # 移动距离x = v0t + 1/2 * a * t^2
        move = v0 * t + 1 / 2 * a * t * t
        # 当前位移
        current += move
        # 加入轨迹
        track.append(round(move))
    return track
```

```python
def get_difference(out1,out2):
    left = 54  # 这个是去掉开始的一部分
    for i in range(left, out1.size[0]):
        for j in range(out2.size[1]):
            # 判断两个像素点是否相同
            pixel1 = out1.load()[i, j]
            pixel2 = out2.load()[i, j]
            threshold = 127
            if abs(pixel1[0] - pixel2[0]) < threshold and abs(pixel1[1] - pixel2[1]) < threshold and abs(
                    pixel1[2] - pixel2[2]) < threshold:
                pass
            else:
                left = i
#     print('缺口位置', left)
    # 减去缺口位移
    left -= 44
    return left
```

```python
def get_csv(all_company):
    import pandas as pd
    company_names_list=[]
    company_urls_list=[]
    company_social_code_list=[]
    company_position_list=[]
    company_hy_list=[]
    company_yggm_list=[]
    company_djzt_list=[]
    company_cbrs_list=[]
    company_zcdz_list=[]
    company_dz_list=[]
    company_person_list=[]
    company_money_list=[]
    company_people_list=[]
    company_day_list=[]
    company_email_list=[]
    company_qylx_list=[]

    for item in all_company:
        if len(item.c_name)>0:
            company_names_list.append(str(item.c_name[0]))
        else:
            company_names_list.append('')
        if len(item.c_url)>0:
            company_urls_list.append(str(item.c_url[0]))
        else:
            company_urls_list.append('')
        if len(item.c_money)>0:
            company_money_list.append(item.c_money[0])
        else:
            company_money_list.append('')
        if len(item.c_position)>0:
            company_position_list.append(item.c_position[0])
        else:
            company_position_list.append('')
        if len(item.c_people)>0:
            company_people_list.append(item.c_people[0])
        else:        
            company_people_list.append('')
        if len(item.c_day)>0:
            company_day_list.append(item.c_day[0])
        else:
            company_day_list.append('')
        if len(item.c_email)>0:
            company_email_list.append(item.c_email[0])
        else:
            company_email_list.append('')
        if len(item.social_code)>0:
            company_social_code_list.append(item.social_code[0])
        else:
            company_social_code_list.append('')
        if len(item.c_qylx)>0:
            company_qylx_list.append(item.c_qylx[0])
        else:
            company_qylx_list.append('')
        if len(item.c_hy)>0:
            company_hy_list.append(item.c_hy[0])
        else:
            company_hy_list.append('')
        if len(item.c_zcdz)>0:
            company_zcdz_list.append(item.c_zcdz[0])
        else:
            company_zcdz_list.append('')
        if len(item.c_cbrx)>0:
            company_cbrs_list.append(item.c_cbrx[0])
        else:
            company_cbrs_list.append('')
        if len(item.c_yggm)>0:
            company_yggm_list.append(item.c_yggm[0])
        else:
            company_yggm_list.append('')
        if len(item.c_djzt)>0:
            company_djzt_list.append(item.c_djzt[0])
        else:
            company_djzt_list.append('')
        if len(item.c_dz)>0:
            company_dz_list.append(item.c_dz)
        else:
            company_dz_list.append('')
    result= pd.DataFrame({'企业名称':company_names_list,'企业链接':company_urls_list,'注册资金':company_money_list,'企业注册位置':company_position_list,
                     '企业代表人':company_people_list,'企业注册日':company_day_list,'企业联系邮箱':company_email_list,'企业社会信用码':company_social_code_list,
                    '企业行业':company_hy_list,'企业注册地址':company_zcdz_list,'参保人数':company_cbrs_list,
                     '员工规模':company_yggm_list,'登记状态':company_djzt_list,'企业所在地址':company_dz_list
                     })
    
    return result
```
